```
# We will import Weather data an plot Temperature by Day
# https://datahub.io/core/global-temp
```

```
# Load ggplot2 package

library(ggplot2)
library(dplyr)
```

```
weatherData = read.csv('https://datahub.io/core/global-temp/r/annual.csv')
```

```

gcagTemp = weatherData[weatherData$Source == 'GCAG',]
```

```
head(gcagTemp)
```

```
hist(gcagTemp$Mean)
grid()
```

```
# Let's compare the two temperature data streams

p = weatherData %>% ggplot( aes(x=Mean, fill=Source)) + geom_histogram( color="#e9ecef", alpha=0.6, position = 'identity') + scale_fill_manual(values=c("#69b3a2", "#404080")) 
    
```
